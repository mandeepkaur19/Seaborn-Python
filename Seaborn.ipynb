```python
vis1 = sns.distplot(df['InternetUsers'], bins=20)
```

#### 2.0.3  If you want only HISTOGRAM then add ‘kde=False’ inside the brackets

```python
vis1 = sns.distplot(df["InternetUsers"], kde=False)
```

##### 2.0.4 If you only want Distribution line then add “hist=False” inside the brackets

```python
vis1 = sns.distplot(df["InternetUsers"], hist=False)
```

> 2. Bi-Variate analysis: Graph using 2 variables. » boxplot
> 3. Multi-variate analysis: plot the graph using more than 2 variables.

#### 3.2 BOX PLOT

##### A box plot,also known as a box and whisker plot,is a graph that summarizes a set numerical data using boxes and lines to show how the data is distributed 

```python
vis2 = sns.boxplot(data=df, x='IncomeGroup', y='BirthRate') #Bivariate
```

##### 3.2.1 In above boxplot
>1. Line haning below box is : MINIMUM-LINE>
2. Line above box is : MAXIMUM_LINE>
3. Line inside the box is : AVERAGE-LINE (Its either MEAN or MEDIAN>)
4. The Diamond point is : OUTLIER == Anomaly detection > It is the data point which is
very far from other observation. > > It will impact more on Machine learning algorithm

##### 3.3 Linear Model Plot

Here you get the REGRESSION LINE and the SCATTER PLOT

```python
vis3 = sns.lmplot(data = df, x='InternetUsers', y='BirthRate')
```

##### 3.3.1 If you want only Scatter-plot then type "fit_reg=False" inside bracket

```python
vis3 = sns.lmplot(data = df, x = 'InternetUsers', y='BirthRate', fit_reg= False)
```

##### 3.3.2 If you want only regression Line then type “scatter=False” inside bracket

```python
vis3 = sns.lmplot(data= df, x='InternetUsers', y='BirthRate', scatter=False)
```

##### 3.3.3 You can differentiate it by adding labels with "hue" ... You can also give different marker styles

for label we use hue in Pandas (seaborn) & legend in NumPy

```python
vis3 = sns.lmplot(data = df, x='InternetUsers', y='BirthRate',hue="IncomeGroup",markers=['o','*','x','d'])
```

##### 3.3.4 You can also use "col" to individual compare data in inplot

```python
vis3 = sns.lmplot(data=df, x='InternetUsers', y='BirthRate', hue="IncomeGroup",col = 'IncomeGroup')
```
